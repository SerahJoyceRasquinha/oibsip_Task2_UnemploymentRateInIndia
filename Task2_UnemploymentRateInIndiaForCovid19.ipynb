```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the dataset
data = pd.read_csv('DataSet/Task_2_Unemployment_FINALdataset.csv')

# Strip leading and trailing spaces from column names, and convert to lowercase
data.columns = data.columns.str.strip().str.lower()

# Rename the columns to align correctly
data = data.rename(columns={
    'date': 'Date',
    'estimated unemployment rate (%)': 'Unemployment Rate (%)',
    'region': 'Region'
})

# Convert Date column to datetime type
data['Date'] = pd.to_datetime(data['Date'])

# Plotting individual bar graphs for each state
for state, state_data in data.groupby('Region'):
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Date', y='Unemployment Rate (%)', data=state_data)
    plt.title(f'Unemployment Rate Over Time in {state}')
    plt.xlabel('Date')
    plt.ylabel('Unemployment Rate (%)')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

plt.figure(figsize=(15, 8))
sns.barplot(x='Date', y='Unemployment Rate (%)', hue='Region', data=data)
plt.title('Aggregate Unemployment Rate Over Time by Region')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
